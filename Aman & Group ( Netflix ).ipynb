```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(data['Monthly Revenue'])
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Country", data=data, order = data['Country'].value_counts().index) 
plt.xticks(rotation='vertical')
plt.title('Most number of users',fontsize=20)    
plt.ylabel('Count',fontsize=25)                 
plt.xlabel('Country',fontsize=25)                 
plt.show()
```

```python
data['Device'].value_counts()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(x="Device",data=data, order = data['Device'].value_counts().index)
plt.xticks(rotation='vertical')
plt.title('Device mostly used',fontsize=30)  
plt.ylabel('Count',fontsize = 25)   
plt.xlabel('Device',fontsize = 25)                      
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Age", data=data, order = data['Age'].value_counts().index) 
plt.xticks(rotation='vertical')
plt.title('Most age group people',fontsize=20)   
plt.ylabel('Count',fontsize=25)                 
plt.xlabel('Age',fontsize=25)                 
plt.show()
```
